```python
large_number = 1e8
small_increment = 1e-8
n_steps = 1000

theoretical = large_number + np.arange(1, n_steps + 1) * small_increment
actual = np.full(n_steps, large_number)
for i in range(1,n_steps):
    actual[i] = actual[i - 1] + small_increment
```

```python
plt.figure(figsize = (10, 6))
plt.plot(range(1, n_steps + 1), theoretical - large_number, label = "Theoretical")
plt.plot(range(1, n_steps + 1), actual - large_number, label = "Actual (Round-off Error)")
plt.title("Round-Off Error Visualization")
plt.xlabel("Number of steps")
plt.ylabel("Increment from Large Number")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(range(1, n_steps + 1), theoretical - actual, label = "True Error")
plt.title("True Error Visualization")
plt.xlabel("Number of steps")
plt.ylabel("Increment from Large Number")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(range(1, n_steps + 1), abs(((theoretical - actual)/theoretical) * 100), label = "True Relative Error")
plt.title("True Relative Error Visualization")
plt.xlabel("Number of steps")
plt.ylabel("Increment from Large Number")
plt.legend()
plt.grid()
plt.show()
```

# *Case 3*

```python
def factorial(x):
    value = 1
    for i in range(1, x + 1):
        value *= i
    return value
```

```python
def taylor_exp(x, n_terms):
    approximation = np.zeros_like(x)
    for n in range(n_terms + 1):
        approximation += x ** n / factorial(n)
    return approximation

x = np.linspace(0 ,2, 100)
exact = np.exp(x)
n_terms_list = [1, 2, 5, 10]

plt.figure(figsize = (10, 6))
for n_terms in n_terms_list:
    approx = taylor_exp(x, n_terms)
    plt.plot(x, approx, label = f"{n_terms} terms")

plt.plot(x, exact, "k--", label = "Exact e^x")
plt.title("Truncation Error in Taylor Series Approximation")
plt.xlabel("x")
plt.ylabel("e^x")
plt.legend()
plt.grid()
plt.show()
    
```

```python
plt.figure(figsize = (10, 6))
for n_terms in n_terms_list:
    approx = taylor_exp(x, n_terms)
    plt.plot(x, exact - approx, label = f"{n_terms} terms")

plt.title("True Error Visualization")
plt.xlabel("x")
plt.ylabel("e^x")
plt.legend()
plt.grid()
plt.show()
    
```

```python
plt.figure(figsize = (10, 6))
for n_terms in n_terms_list:
    approx = taylor_exp(x, n_terms)
    plt.plot(x, abs(((exact - approx) / exact) * 100), label = f"{n_terms} terms")

plt.title("True Relative Error Visualization")
plt.xlabel("x")
plt.ylabel("e^x")
plt.legend()
plt.grid()
plt.show()
    
```
